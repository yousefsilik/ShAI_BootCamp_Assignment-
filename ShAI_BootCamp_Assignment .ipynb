```python

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/Salaries.csv")
job_title_counts = df['JobTitle'].value_counts()
top_n = 10
top_job_titles = job_title_counts.head(top_n)
other_job_titles = job_title_counts.iloc[top_n:]
combined_job_titles = pd.Series({'Other': other_job_titles.sum()})
final_job_title_counts = top_job_titles.append(combined_job_titles)
plt.figure(figsize=(8, 8))
plt.pie(final_job_title_counts, labels=final_job_title_counts.index, autopct='%1.1f%%', startangle=90
        , colors=['red', 'orange', 'yellow', 'green', 'blue','black','grey'])
plt.show()

```

```python
import pandas as pd
df = pd.read_csv("/Salaries.csv")
grouped_data = df.groupby('JobTitle')
summary_statistics = grouped_data['TotalPay'].agg(['mean', 'median', 'std', 'count'])
print(summary_statistics)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/Salaries.csv")
numerical_column = 'BasePay'
correlation_coefficient = df['TotalPay'].corr(df[numerical_column])
print(f"Correlation between TotalPay and {numerical_column}: {correlation_coefficient:.2f}")
plt.figure(figsize=(10, 6))
plt.scatter(df[numerical_column], df['TotalPay'], alpha=0.5)
plt.title(f'Scatter Plot: TotalPay vs {numerical_column}')
plt.xlabel(numerical_column)
plt.ylabel('TotalPay')
plt.grid(True)
plt.show()
```
